```python
"""
========
Barchart
========

A bar plot with errorbars and height labels on individual bars
"""
import numpy as np
import matplotlib.pyplot as plt

N = 5
women_means = (5.88,5.12,4.62,7.76,6.62)
men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='r')

men_means = (6.64,6.59,6.45,6.32,5.09)
women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Current vs Desired')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('confort fin', 'proxim.dom.', 'Flexwork', 'Int.Stim', 'Domain motiv'))

ax.legend((rects1[0], rects2[0]), ('Current', 'Desired'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,round(height,2),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
