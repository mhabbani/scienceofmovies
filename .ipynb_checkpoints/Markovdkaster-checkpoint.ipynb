```python
import numpy as np
import pandas as pd
from nltk import word_tokenize,sent_tokenize
from nltk.util import ngrams

dfm = pd.read_csv("./data/micro_critique.csv", sep="\t",encoding="utf-8-sig", error_bad_lines=False, low_memory=False)
```

```python
# load dataset ratings (dfr) and movies (dfm)
dfr = pd.read_csv("./data/vod_ratings.csv", sep="\t",encoding="utf-8-sig", error_bad_lines=False, low_memory=False)
# merge ratings & critics pour be able so select good & bad critics
dfr = dfr.rename(columns = {'value':'rating','object_id':'film_id'})
df = pd.merge(dfr, dfm, on=['user_id','film_id'])

df[:5]
```

```python
#création des n_grames
def word_grams(words, size):
    s = []
    for ngram in ngrams(words, size):
        s.append(' '.join(str(i) for i in ngram))
    return s

print (word_grams(df.loc[5,"content"].split(' '),2))
```

```python
#tire un mot aléatoire avec une proba accrue fonction de sa fréquence
def getrandword(dic):
    freq=dic["freq"]
    nb = np.random.randint(freq)
    word="notfound."
    for k in dic.keys():
        if k!="freq":
            nb = nb-dic[k]
            if nb<0:
                return k;
    return word;

#vérifie s'il y a une ponctuation à la fin de la chaine de caractères
def contains_punctuation(chaine):
    if chaine[len(chaine)-1]==".":
        return True;
    if chaine[len(chaine)-1]=="!":
        return True;
    if chaine[len(chaine)-1]=="?":
        return True;
    return False;
```

```python
dft = df[df['rating']<2]
dft.shape
```

```python
#création du corpus de data et de 1ers mots
data = ""
first_words={}
first_words['freq']=0
for i in range(len(dft)):
    critique = dft.iloc[i,dft.columns.get_loc('content')]
    if i%10000==0:
        print(i)
    data = data + str(critique) + " "
    critique_split=str(critique).split(' ')
    if len(critique_split)>0:
        first_word=critique_split[0]
        if first_word in first_words:
            first_words[first_word]+=1
            first_words['freq']+=1
        else:
            first_words[first_word]=1
            first_words['freq']+=1
print(first_words)
```

```python
import pickle
pickle.dump(data, open( "critiques_1.p", "wb" ))
pickle.dump(first_words, open( "first_words_1.p", "wb" ))
#data = pickle.load( open( "critiques.p", "rb" ))
```

```python
#creation d'un dictionnaire avec clé : mot 1 puis mot 1.1 : freq (mot1.1), mot 1.2 : freq (mot1.2), etc.
dic_corpus = {}
corpus = word_grams(data.split(' '),2)
for i in range(len(corpus)):
    mot1 = corpus[i].split(' ')[0]
    mot2 = corpus[i].split(' ')[1]
    #Création d'une entrée pour le mot1 (état présent)
    if mot1 in dic_corpus:
        dic_corpus[mot1]["freq"]+=1
    else:
        dic_corpus[mot1]={}
        dic_corpus[mot1]["freq"]=1
    #MAJ de la fréquence du mot2 pour l'état présent mot1
    dic_corpus[mot1][mot2] = dic_corpus[mot1].get(mot2, 0) + 1
```

```python
lenght = 12
how_many = 10

for h in range(how_many):
    #Choix du 1er mot
    current_word = getrandword(first_words)

    #Construction chaine de markov
    critique = current_word
    i=0
    while i<lenght:
            i+=1
            current_word = getrandword(dic_corpus[current_word])
            critique += " "+current_word

            #on continue tant qu'on n'est pas à une fin de phrase
            if i==lenght:
                if contains_punctuation(current_word) == False:
                    i-=1

    print (str(h)+':'+critique)
```
